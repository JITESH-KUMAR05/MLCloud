```python
import pandas as pd
messages = pd.read_csv('spam.csv',sep=',' , usecols=[0,1] , encoding='latin1' , header=0, names=['label','message'])
```

```python
messages
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer
import re
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(messages)):
    review = re.sub('^a-bA-B',' ',messages['message'][i])
    review = review.lower()
    review  = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
## for binary we can keep binary=True
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
## Independent Feature
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
                    formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
cv.vocabulary_
```

```python
## output Feature
y = pd.get_dummies(messages['label'])
```

```python
y
```

```python
y = y.iloc[:,0].values
```

```python
## Train test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detection_model = MultinomialNB().fit(X_train,y_train)
```

```python
spam_detection_model
```

```python
y_predict = spam_detection_model.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_predict,y_test)
```

```python
print(classification_report(y_test,y_predict))
```
